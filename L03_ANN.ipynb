 * They are inspired by the brain, which consists of a network of brain cells (neurons)
 * Mathematically a neuron is a concatenation of a weighted function $h()$, followed by a *non linearity* $g()$.

## Activation Functions

Most commonly used activation functions

 * Sigmoid: $\sigma(x)=\frac{1}{1+e^{-x}}$

 * Tanh (Hyperbolic Tangent): $tanh(x) = \frac{e^x-e^{-x} }{e^x+e^{-x}} $

 * ReLu (Rectified Linear Unit): $ReLU(x) = \begin{cases} 0 & x < 0 \\ x & x \ge 0 \end{cases} = max(0,x) $ 

![Activation Functions](https://drive.google.com/uc?export=download&id=13A7SZceUcla1HzpARDlzc9-HMCVArPMB)
<!-- https://drive.google.com/file/d/13A7SZceUcla1HzpARDlzc9-HMCVArPMB/view?usp=sharing --> 

```python
# Code to draw the Sigmoid function
import matplotlib.pyplot as plt
import numpy as np

def showActivationFunctions():
  xlb=-8 # x Left Bound
  xrb=8  # x Right Bound

  yub=2.0  #y Upper Bound
  ylb=-1.5 # y Lower Bound
  
  sigmoid = lambda x: 1 / (1 + np.exp(-x)) 
  tanh = lambda x: (np.exp(x)-np.exp(-x))/(np.exp(x)+np.exp(-x))
  relu = lambda x: [max(0,el) for el in x]
  xVec = np.linspace(xlb, xrb, 1200) # start, stop, n. of points

  plt.clf()
  plt.title("Activation Functions Representation")
  
  fname2lineStyle={
      'Sigmoid':'-.', 
      'Tanh':'--', 
      'ReLU':':'
  }

  for fname, funct in {
      'Sigmoid':sigmoid, 
      'Tanh':tanh, 
      'ReLU':relu
    }.items():
    
    plt.plot(
        xVec, 
        funct(xVec), 
        label=fname, 
        ls=fname2lineStyle[fname] # ls is line style
        )
  
  plt.legend() #Creates a legend
  plt.xlabel("x axis")
  plt.ylabel("y axis")
  
  plt.axhline(0, color='gray', ls=':') # x axis
  plt.axvline(0, color='gray', ls=':') # y axis
  
  plt.ylim(ylb,yub)
  plt.yticks(np.arange(ylb, yub, step=0.5)) 
  plt.xticks(np.arange(xlb, xrb, step=1)) 
  #plt.savefig('activationFunctions.png') #uncomment to save on disk. Also call this method before plt.show() to obtain a high quality figure
  plt.show()

showActivationFunctions()
```

All hidden layers typically use the same activation function. 

The output layer 

* it will typically use a different activation function from the hidden layers

### ReLU Pros and Cons
The rectified linear activation function, or ReLU activation function, is perhaps the most common function used for hidden layers.


ReLU usage is common because 
* it is both simple to implement, 

* easy to compute (its derivative is very simple), and 

* it is less susceptible to vanishing gradients (this topic will be addressed later) w.r.t. Sigmoid and Tanh, although it can suffer from other problems like saturated or dead units.


### Sigmoid Advices

Also called the logistic function.

When using the Sigmoid function for hidden layers, it is a good practice to scale input data to the range 0-1 (e.g. the range of the activation function) prior to training.

### TanH Advices

When using the TanH (Hyperbolic Tangent)  for hidden layers, it is a good practice to scale input data to the range -1 to 1 (e.g. the range of the activation function) prior to training.




#### (How to choose) the Activation Function for Internal Layers

Both the sigmoid and Tanh functions can are affected by the so-called vanishing gradients problem.

Hence, Modern (deep) neural network models, such as MLP (Multi Layer Perceptron) and CNN (Convolutional Neural Network), will make use of the ReLU activation function.


Nevertheless, Recurrent networks still commonly use Tanh or sigmoid activation functions, or even both. For example, the LSTM commonly uses the Sigmoid activation for recurrent connections and the Tanh activation for output.


Generally speacking ... the hyperbolic tangent activation function typically performs better than the logistic sigmoid

#### (How to choose) the Activation Function for Output Layers

There are 3 activation functions most frequently used for output layer:

* Linear Activation
* Logistic (Sigmoid)
* Softmax

This is not an exhaustive list, but they are the most commonly used.

##### Linear

The linear activation function is also called identity or no activation.

This is because the linear activation function returns the input value directly.

It is used whent the ANN should perform regression acitivities.

Target values used to train a model with a linear activation function in the output layer are typically scaled prior to modeling using one of the following:

* **min max scaler** i.e., $\; y = \frac{x \; - \; min}{max \; - \; min}$

* **normalization** i.e., $\; y = \frac{x \; - \; mean}{standard\_deviation}$

##### Sigmoid

Sigmoid is used when the ANN should be a binary classifier.

Target labels used to train a model with a sigmoid activation function in the output layer must have the values 0 or 1.

If there are two or more mutually inclusive classes (**multilabel classification**), then your output layer will have one node for each class and a sigmoid activation function is used.

##### Softmax

The softmax function is used when the ANN should be a **Multiclass Classifier**.

A multiclass classifier always select one (and only one) label among a set of several ones.

![](https://drive.google.com/uc?export=download&id=13W5RYnk8RSWmknY4AOqWQ5RF6DogU6SG)
<!-- https://drive.google.com/file/d/13W5RYnk8RSWmknY4AOqWQ5RF6DogU6SG/view?usp=sharing -->

Softmax is a mathematical function that converts a vector of numbers into a vector of probabilities, where the probabilities of each value are proportional to the relative scale of each value in the vector.

An ANN using softmax 
* is configured to **output N values** (one for each class in the classification task), and 
* the softmax function normalizes the outputs, converting them into **values that sum to one**. 
* Each value in the output of the softmax function is interpreted as the **probability of membership** for each class.


Suppose you have a multiclass classification problem where the classifier should choose one class among $c_i \in [c_1, ..., c_n]$. 
Suppose that the output layer has the vector $Z=[z_1, ..., z_n]$ in input (the size has to be equal to the number of classes).

The softmax  function 
\begin{align} 
\sigma :\mathbb{R}^K\to (0,1)^K \text{ with }  k>1
\end{align}

is defined by the formula

\begin{align} 
\sigma (\mathbf {z} )=\left[{\frac {e^{z_{i}}}{\sum _{j=1}^{K}e^{z_{j}}}} {\text{ , for }}i=1,\dotsc ,K \ \ \right]
\end{align}

When Softmax is used in the output layer, it is suggested to use as Target Labels  
* 1 for the target class, and 
* 0 for all other classes.

# Backpropagation

Consider a linear model having 
* several input, 
* absence of non-linearity
* several outputs (this is a little bit different from a neuron)

where the output values $y_j$ are linear combinations of the input variables $x_i$ so that $y_j = \sum_i w_{j,i} \cdot x_i$

Then, 

* given a pair $(X, T)$ where $X$ is an input vector and $T$ is the desired output vector.  Specifically, 
 
 * $X = [x_1, ..., x_{|X|}]$ 
 * $T = [t_1, ..., t_{|T|}]$


* we can compute an **Error Function** $EF = \frac{1}{2} \sum_{j} (y_j - t_j)^2$ where
   * $y_j = \sum_{i=1}^{|X|} w_{j,i} \cdot x_i$ 
   * $w_{j,i}$ reads: the weight connecting the $i^{th}$ input to the $j^{th}$ output

We can try to find the optimal $w_{\cdot,\cdot}$ values (which minimize the error) using gradient descent

We need to compute the derivative

$\frac{\partial EF}{\partial w_{j,i}} = \frac{\partial \frac{1}{2} \sum_{j} (y_j - t_j)^2}{\partial w_{j,i}} $

In this setting, we can have a very large set of $w_{j,i}$ weights. Computing from scratch all the derivates can take a long time.

Fortunately, there is a workaround

We recall that 

* $\frac{\partial f(g(x))}{\partial x} = f'(x) \cdot  g'(x)$

* $y_j = \sum_{i=1}^{|X|} w_{j,i} \cdot x_i$  

* $\frac{\partial y_j}{\partial w_{j,i}} = x_i$

Hence

$\frac{\partial EF}{\partial w_{j,i}} = \frac{\partial \frac{1}{2} \sum_j (y_j - t_j)^2}{\partial w_{j,i}} =  (y_j - t_j) \cdot x_i$

This result can be interpreted as follows: 

$\frac{\partial EF}{\partial w_{j,i}}$  can be the result of a simple *local* computation involving the product of both 

* an **error signal** $y_j - t_j$ representing the output error 

* the corresponding input value $x_i$

We can rewrite the output error as follows:

$\delta_j = y_j - t_j$

hence,

$\frac{\partial EF}{\partial w_{j,i}} = \frac{\partial \frac{1}{2} \sum_{j=1}^{|T|} (y_j - t_j)^2}{\partial w_{j,i}} =  (y_j - t_j) \cdot x_i = \delta_j \cdot x_i$

We will see that this is a common pattern.

## Now switching to ANN. Suggested Approach

In a feed forward neural network, 
* weights are random initialized (they cannot be initialized using a constant value)

* The network output is computed using the weights (forward pass)

* The weights are updated using gradient descent i.e., 

$
 w_{j,i}^{new} = w_{j,i} - \frac{\partial EF}{\partial w_{j,i}}  \cdot learning\_rate
$

##Partial derivatives of an internal layer neuron

In a feed forward neural network, each neuron computes a weighted sum of its input, as follows:

$a_j = \sum_i w_{j,i} \cdot x_i$

For the sake of simplicity, we are not considering the bias term.

The neuron final output is computed using a nonlinear activation function $h(\cdot)$ e.g.,

$z_j = h(a_j)$

For each element in the input vector we can compute the neuron output by applying the above formula

This process is often called **forward propagation** because it can be regarded as a forward flow of information through the network

We can still use gradient descent to find the optimal $w_{\cdot,\cdot}$ values.

First we note that EF dependes on the weight $w_{j,i}$ only via the summed input $a_j$ to unit $j$. We can therefore apply the chain rule for partial derivatives to give

$
\frac{\partial EF}{\partial w_{j,i}} = \frac{\partial EF}{\partial a_j} \cdot \frac{\partial a_j}{\partial w_{j,i}}
$

We now introduce a useful notation

$\delta_j = \frac{\partial EF}{\partial a_j} $

where $\delta$'s are often referred to as **errors** for reasons we shall see shortly.

We recall that 

$\frac{\partial a_j}{\partial w_{j,i}} = z_i$

hence the partial derivative can be rewritten as 

$
\frac{\partial EF}{\partial w_{j,i}} = \delta_j \cdot z_i
$

For the output units we have

$\delta_j = y_j - t_j$

while, to obtain the $\delta$'s for hidden units, we again make use of the chain rule for partial derivatives

$
\delta_j = \frac{\partial EF}{\partial a_j} = \sum_k \frac{\partial EF}{\partial a_k} \cdot \frac{\partial a_k}{\partial a_j}
$

where the sum runs over all units $k$ to which unit j sends connections. The arrangiement of units and weights is issustrated in the figure below.

![Backpropagation on hidden units](https://drive.google.com/uc?export=download&id=13SpWlb2UY7lkqz3UguEH4D22mARLELRK)
<!-- https://drive.google.com/file/d/13SpWlb2UY7lkqz3UguEH4D22mARLELRK/view?usp=sharing -->

Using previous formulae we can rewrite 

$
\delta_j = h'(a_j) \sum_k w_{k,j} \cdot \delta_k
$

which tells us that the value of $\delta$ for a particular hidden unit can be obtained by propagating the $\delta$'s backwards from units higher up in the network.

The backward propagation procedure can therefore be summarized as follows

1. Apply an input vector $X$ and forward propagate through the network using $a_j = \sum_i w_{j,i} \cdot x_i$ and $z_j = h(a_j)$ to find the output of all the hidden and output neurons

2. Compute the $\delta_k$ for all the output units

3. Backpropagate the $\delta$'s using $\delta_j = h'(a_j) \sum_k w_{k,j} \cdot \delta_k$ to obtain $\delta_j$ for each hidden unit in the network

4. Use $
\frac{\partial EF}{\partial w_{j,i}} = \delta_j \cdot z_i
$ to compute the required derivatives

5. Update the $w$ values using the computed derivatives

We assumed a training set composed by a simple $X$ vector. The above considerations can be generalized over multiple input vectors.

###Random initialization

If the same value were used to initialize the weights $w$ of an Artificial Neural Network ...

... Backpropagation will change all the weights in the same way. 

As a result, the ANN training will fail.

It is a good habit to initialize the $w$'s using random values.

### Vanishing Gradient Descent

Unfortunately, deep neural networks (i.e., neural networks having several internal layer) might bring the error $\delta$ very close to 0 in the very internal layer. This problem, known as **Vanishing Gradient Descent** might make an Artificial Neural Network unsuitable to learn.

###Stochastic Gradient Descent

It can be regarded as a stochastic approximation of gradient descent optimization, since it replaces the  gradient calculated from the entire data set, by an estimate thereof (calculated from a randomly selected subset of the data). This approach reduces the computational effort required to train an Artificial Neural Network.

Prevent Overfitting in CNN

* **Max Pooling**. It partitions the input image/data into a set of non-overlapping rectangles. For each such sub-region, the maximum feature value is selected. 

 * It is a form of non-linear downsampling 
 * The intuition is that the exact location of a feature is less important than its rough location
  

* **Drop Out**. During training only a limited subset of features is propagated to the next layer i.e., neuron subsets are randomly turned off

 *  It prevents he subsequent layers to become too lazy due to overwhelming information
 * During prediction, all features are propagated to maximize the classification results i.e., no neuron subset is turned off

#Autoencoders

An autoencoder neural network is a neural netwok trying to produce output equal to the inputs, i.e., $y=x$

It seems a particularly trivial function to be trying to learn ...

... but by placing constraints on the network, such as by limiting the number of hidden units, interesting structures about the data can be discovered (eg., meaningful low dimension features / latent variables)

![Autoencoder](https://drive.google.com/uc?export=download&id=13gUIZWLHyd90lUqQh-BZ2Ftu4eo4tgy0)
<!-- https://drive.google.com/file/d/13gUIZWLHyd90lUqQh-BZ2Ftu4eo4tgy0/view?usp=sharing -->

The desired output is a byproduct of the autoencoder: the desired output are the central layer neuron's output.
   
* Since the internal layer is forced to represent the same information in less space, it is forced to abstract
* e.g., instead of storing all the pixels of a square, it will store an *abstract concept* like *"there is a square"*


The autoencorder *learns* the internal features (latent variables) during an unsupervised process. No need for labelled data.

The Autoencoder byproducts can be used to initialise the layers of a Convolutional Neural Network, to speed up the  training and improve the classification performances

# Embeddings (e.g., Word2Vec)

The intuition behind Word2Vec (or in general Word Embeddings) is that some useful information can be extracted as a by-product of a machine learning process.

Given a word $w$ and its context $k$ (a set of $m$ words nearby $w$), 

The context $k$  can be used as a feature for predicting the word $w$. 

The last problem can be viewed as a neural network learning problem where the representation of $k$ (i.e., $m$ context words) is fed into a network trained to predict the representation of $w$. 

A similar (but reversed problem) is the Skip-n-gram model i.e., to train a neural network to predict the representation of the context $k$ (i.e., the $m$ context words) from *w*

Word2Vec embeddings related to words have nice properties, e.g., 

* $[Madrid] - [Spain] + [France] \approx [Paris] $
* $[King] - [Man] + [Woman] \approx [Queen] $
