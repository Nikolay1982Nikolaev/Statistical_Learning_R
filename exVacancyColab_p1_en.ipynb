```


##########################
# Unfortunately, the dataset is unbalanced
# Balancing by undersammpling
def underSample2Min(df, labelName):
    ''' The dataset is undersampled so that all label groups will have the same size,
        corresponding to the (original) minimal label set.
        The parameter labelName is the DataFrmae column hosting the labels'''
        
    vc = df.loc[:,labelName].value_counts() # Counting label frequencies
    lab2freq = dict(zip(vc.index.tolist(), vc.values.tolist()))
    #print(lab2freq) # if you want to see lab2freq, please uncomment this command
    #print(min(lab2freq.values()))
    minfreq = min(lab2freq.values())
    #print(minfreq)
    idxSample=[]
    for selectedLabel, actualFreq in lab2freq.items():
        selIndexes=df.loc[df.loc[:,labelName]==selectedLabel, :].sample(n=minfreq).index.tolist()
        idxSample+=selIndexes
    idxSample.sort()
    #print(type(idxSample), idxSample)
    #print(list(df.index)[:5])
    
    df2 = df.loc[idxSample, :]
    #print(len(idxSample), df2.shape);exit()
    df2 = df2.reset_index() # otherwise missing index may cause problem
    return df2

df3 = underSample2Min(df2, 'CategoryID')

# Let's look again to the category sizes
print('Cat. sizes')
print(df3.loc[:,'CategoryID'].value_counts())
```

'''

# Text Processing and Machine Learning

Text data requires special preparation before using machine learning algorithms. Some details are provided in the lines below.


## Bag-of-Words Model

We may want to perform classification of documents (whereas a document is a set of strings), so each document is an âinputâ and a class label is the "output". Machine Learning algorithms take vectors of numbers as input, therefore we need to convert documents to fixed-length vectors of numbers.

A simple and effective way for dealing with text documents in machine learning is called the Bag-of-Words approach, or BoW.

The approach is simple in that it throws away all of the order information in the words and focuses on the occurrence of words in a document.

This can be done by assigning each word to a unique place in a fixed lenght-vector. The vector length will be the length of the vocabulary of known words. The value in each vector position could be filled with the frequency of the corresponding word in the encoded document.

This is the bag of words approach, where we are only concerned with encoding schemes that represent the degree to which words are present in encoded documents without any information about order.

There are many ways to extend this simple method, both by better clarifying what a âwordâ is and in defining what to encode about each word in the vector.

To recap what we have just read, 
- the text must be parsed to identify single words, this process is called tokenization. 

- Then, the words need to be encoded as integers or floating point values for use as input to a machine learning algorithm. The latter process is called feature extraction (or vectorization).


## **Scikit-learn** library

The [CountVectorizer](http://scikit-learn.org/stable/modules/generated/sklearn.feature_extraction.text.CountVectorizer.html) in the scikit-learn library provides a simple way to:

- encode documents using the Bag of Word approach

- tokenize a collection of text documents

- build a vocabulary of known words

- encode new documents using that vocabulary

The CountVectorizer should be used as follows:

- create an instance of the CountVectorizer class.

- call the fit() function  to learn a vocabulary from a set of documents.

- call the transform() function on a set of document to produce the document-term-matrix (i.e., to encode the documents).

An matrix is returned, whereas each (row) vector has the  length of the entire vocabulary, and each (row vector) element is an integer count for the number of times the corresponding word appeared in the document.

Because data contain a lot of zeros, they are stored in a sparse matrix data structure. They can be transformed into dense arrays before printing by calling the .toarray() method.

Below is an example of using the CountVectorizer to tokenize, build a vocabulary, and then encode a document.

'''

```
######
# Some examples using the CountVectorizer
# Some code is commented using multi line strings, 
# if you like you can uncomment 
# (by dropping the ''' both at the beginning and at the end)
# and then execute the code
 
# In python, multi-line strings are used as multi-line comments 

#importing the CountVectorizer library
from sklearn.feature_extraction.text import CountVectorizer

# list of text documents. For the sake of simplicity we have only 2 documents below.
text = ["The quick brown fox jumped over the lazy dog.", 
        "The quick brown fox"
        ]

# Instantiating the CountVectorizer() class, so that it can be used to build the tf-matrix
vectorizer = CountVectorizer()

# proces the text (e.g., tokenize) and build the vocabulary
# In a vocabulary each word is unique (no repetition)
vectorizer.fit(text)

print('Vocabulary')
print(vectorizer.get_feature_names_out()) 
# ['brown', 'dog', 'fox', 'jumped', 'lazy', 'over', 'quick', 'the']
print()

# Now we compute the BoW Matrix. 
# We can work on different documents too!
# But the vocabulary won't change (i.e., new terms are not added)
documentTermMatrix = vectorizer.transform(text)

print("Let's have a look to the output.")
print('documentTermMatrix shape')
print(documentTermMatrix.shape) 
# (2, 8)
print()

print('documentTermMatrix is of type: ')
print(type(documentTermMatrix)) 
#Â scipy.sparse.csr.csr_matrix are sparce matrix
print()

print("and its content is as follows (each matrix element")
print("is represented using the form ...")
print("  (x,y) value")
print("... where x,y are the cell coordinates, and value is the cell content):")
print(documentTermMatrix)
# ...
#  (0, 0)	1
#  (0, 1)	1
#  (0, 2)	1
#  (0, 3)	1
#  (0, 4)	1
#  (0, 5)	1
#  (0, 6)	1
#  (0, 7)	2
#  (1, 0)	1
#  (1, 2)	1
#  (1, 6)	1
#  (1, 7)	1
print()

# .todense() turns the matrix into a dense one. 
# A dense matrix is more print friendly
print('Showing the corresponding dense matrix')
print('Dense matrix type') 
print(type(documentTermMatrix.toarray())) 
# <class 'numpy.ndarray'> is a n-dimensional array
print('Print Output')
print(documentTermMatrix.todense())  
# [
#    [1 1 1 1 1 1 1 2], 
#    [...]
# ] 
```

```
##########################
# Transforming the dtm into a pandas DataFrame
# numpy.ndarray and scipy.sparse.csr.csr_matrix are data types primarily used for computation
# They cannot store meta-data, while pandas DataFrames can have meta-data with data

# get feature names
columnNames = vectorizer.get_feature_names()

# using feature names as column names in the pandas DataFrame
dtmdf = pd.DataFrame(documentTermMatrix.todense(), columns=columnNames)
display(dtmdf.head())
print('Please, ignore the first column which is the dataframe index')
```

The encoded vectors can then be used directly with a machine learning algorithm.

```
######
# Simple Pipeline
# tockenization and stop words filtering is performed using 
# CountVectorizer built-in functions

# Let's build a pipeline for preprocessing and classifying job vacancy titles
cp = Pipeline([  
   ('vectorizer', CountVectorizer()),
   ('classifier', LogisticRegression() ), # LinearSVC is another classifier you can try
                           ])

# We collect all the configuration parameters inside a single data structure. 
# Each parameter is identified by the string pipelineComponentName__ParamName.
# The pipeline component name and the param name are separated 
# by 2 underscore "_"
clsfParams = {
   'classifier__C': 0.001,  # C is equivalent to 1/lambda
   'vectorizer__stop_words':'english', #Â articles are discardedÃ
   'vectorizer__ngram_range': (1,1), #only unigrams are considered. (1,2) is for both unigrams and bigrams, (1,3) is for unigrams bigrams and trigrams, ...
}

# The parameters are set NOW, after pipeline creation
# This is strange at the very beginning,
# but later you will apreciate it
cp.set_params(**clsfParams) 
cp.fit(xTrain, yTrain)

# Performing the prediction
yPred=cp.predict(xTest)

#Computing the classification report
print('Classification report')
clasRepSt01 = classification_report(yTest,yPred)
print(clasRepSt01)

#Computgin accuracy too
print('Accuracy')
print(accuracy_score(yTest,yPred))

# What will be the accuracy, in your opinion?












#
```
